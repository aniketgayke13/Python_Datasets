## Correlations-

 - calculating the co-relation
 - co-relation lies in between -1 to +1 
 - it is a negative co-relation as wieght increases and milage decrease

```python
# importing the library
import pandas as pd
```

```python
# importing the data set
mtcars = pd.read_csv(r"O:\IMARTICUS LECTURES\Python\Pandas Data Set\Mtcars.csv")
mtcars.head()
```

```python
# getting the correlation between numeric columns

mtcars.mpg.corr(mtcars.wt) 
```

```python
# where as 
mtcars.wt.corr(mtcars.mpg)

# as we know c0-relation between corr(A,B) = corr(B,A)
# i.e, mpg,wt = wt,mpg
```

```python
mtcars.corr()

# this is known as your co-relation matrix
# as we can see in corelation matrix diagonal elemnts will always be 1
```
